```python
#Lecture3
#Issue: What cost(W) look like?

#Gradient decent algorithm 경사 하강 알고리즘
#Object : Minimize cost function
#It can be applied to more general function: cost(w1,w2,w3....)


import tensorflow as tf
import matplotlib.pyplot as plt
x=[1,2,3]
y=[1,2,3]

w = tf.placeholder(tf.float32)
H = w*x

cost = tf.reduce_mean(tf.square(H-y))
sess = tf.Session()

sess.run(tf.global_variables_initializer())

w_val =[]
cost_val=[]

for i in range(-50,100):
    curr_cost,curr_w=sess.run([cost,w],feed_dict={w:i*0.1})
    w_val.append(curr_w)
    cost_val.append(curr_cost)
    #print(curr_cost,curr_w)
    
plt.plot(w_val,cost_val)





```

```python
import tensorflow as tf


x_data = [1,2,3]
y_data = [1,2,3]

w = tf.Variable(tf.random_normal([1]),name="weight")
x = tf.placeholder(tf.float32)
y = tf.placeholder(tf.float32)


H = x*w

cost = tf.reduce_sum(tf.square(H-y))

learning_rate = 0.1
gradient = tf.reduce_mean((w*x-y)*x)
descent = w - learning_rate*gradient
update = w.assign(descent)

sess = tf.Session()

sess.run(tf.global_variables_initializer())

for step in range(100):
    sess.run(update,feed_dict={x:x_data,y:y_data})
    print(step,sess.run(cost,feed_dict={x:x_data,y:y_data}),sess.run(w))



```

```python
# 미분

x = [1,2,3]
y = [1,2,3]
w = tf.Variable(5)
H = x*w

cost = tf.reduce_mean(tf.square(H-y))


optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.1)
train = optimizer.minimize(cost)

sess=tf.Session()
sess.run(tf.global_variables_initializer())

for step in range(100):
    print(step,sess.run(w))
    sess.run(train)

```

```python
X = [1, 2, 3]
Y = [1, 2, 3]

# Set wrong model weights
W = tf.Variable(5.0)

# Linear model
hypothesis = X * W

# cost/loss function
cost = tf.reduce_mean(tf.square(hypothesis - Y))

# Minimize: Gradient Descent Magic
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.1)
train = optimizer.minimize(cost)

# Launch the graph in a session.
sess = tf.Session()
# Initializes global variables in the graph.
sess.run(tf.global_variables_initializer())

for step in range(100):
    print(step, sess.run(W))
    sess.run(train)
```

```python
import tensorflow as tf


x= [1,2,3]
y= [1,2,3]

w = tf.Variable(5.)

H = x*w

gradient = tf.reduce_mean(tf.square(w*x-y)*x)*2

cost = tf.reduce_mean(tf.square(H-y))
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01)

gvs = optimizer.compute_gradients(cost)
####기울기에 계산 추가 생략

apply_gradient = optimizer.apply_gradients(gvs)

sess = tf.Session()
sess.run(tf.global_variables_initializer())


for i in range(100):
    print(i,sess.run([gradient,w,gvs]))
    sess.run(apply_gradient)





```
