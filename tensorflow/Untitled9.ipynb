```python
#lecture 2

#Linear Regression using tensorflow

import tensorflow as tf

# H(x) = Wx + b
# Issue: Which hypothesiss(model) is better?
# Cost function (= Loss function)

# Cost = H(x) - y  (*y = Wx +b)

# How fit the model
# cost = 1/m * sum of (H(x)-y)^2 (=mse)
# Object : minimize cost(W,b)


```

```python
x_train =[1,2,3,4,5]
y_train =[1,2,3,4,5]


#Variable * tensorflow가 자체적으로 사용하는 변수 trainable value
w = tf.Variable(tf.random_normal([1]),name='weigh')
b = tf.Variable(tf.random_normal([1]),name='bias')

H = x_train*w+b


#reduce_mean:
#t = [1,2,3,4,5]
#tf.reduce_mean(t)  ==> 3

#cost function
cost = tf.reduce_mean(tf.square(H-y_train))

#Gradient Descent
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01)
train = optimizer.minimize(cost)

#Launch the graph in a session
sess = tf.Session()

#Initializes global variables in the grapth
#Variable을 사용하기 위해서는 반드시 grobal variable initializer 필요!!
sess.run(tf.global_variables_initializer())

#Fit the line
for step in range(2001):
    sess.run(train)
    if (step % 20 ==0):
        print(step,sess.run(cost),sess.run(w),sess.run(b))
    
```

```python
#Placeholders (만들어진 모델이 값넘기기 가능)
#초기 학습전 random value 지정
w = tf.Variable(tf.random_normal([1]),name='weigh')
b = tf.Variable(tf.random_normal([1]),name='bias')


# x,y shape 정의
x = tf.placeholder(tf.float32,shape=[None])
y = tf.placeholder(tf.float32,shape=[None])

# tf.Variable 사용을 위한 초기화
sess.run(tf.global_variables_initializer())

# H(x) 정의
H = x*w+b

# costfunction 정의
cost = tf.reduce_mean(tf.square(H-y))

# Minimize(object function) 방법 정의
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01)
train = optimizer.minimize(cost)



#모델 학습
for step in range(2001):
    cost_val, w_val, b_val, _ = \
        sess.run([cost, w,b, train],
                 feed_dict={x: [1, 2, 3,5,6], y: [1.2, 2, 3,7,10]})
    if step % 200 == 0:
        print(step, cost_val, w_val, b_val)

#prediction
print(sess.run(H, feed_dict={x: [5,10,20]}))

```
